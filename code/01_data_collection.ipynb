# 01_data_collection notebook

**the objective of this notebook is to collect data from California DMV website.**

## 1. Download disgengagement reports (.csv)

Go to this website:  https://www.dmv.ca.gov/portal/vehicle-industry-services/autonomous-vehicles/disengagement-reports/

Collect disengagement reports in .csv format for 2023, 2022, and 2021 and save in /data folder.

## 2. Imports

```python
import pandas as pd
import numpy as np
```

## 3. Create Dataframes

```python
# create a dataframe for each year California DMV disengagement reports
reports2023 = pd.read_csv('../data/2023-Autonomous-Vehicle-Disengagement-Reports-CSV.csv', encoding='latin1')
reports2023.head()
```

```python
reports2022 = pd.read_csv('../data/2022-Autonomous-Vehicle-Disengagement-Reports-CSV.csv', encoding='latin1')
reports2022.head()
```

```python
reports2021 = pd.read_csv('../data/2021-Autonomous-Vehicle-Disengagements-Reports-CSV.csv', encoding='latin1')
reports2021.head()
```

## 4. Export CSVs

export the dataframes as new csv files for data preparation and analysis later.

```python
reports2023.to_csv('../data/reports2023.csv', index=False)
reports2022.to_csv('../data/reports2022.csv', index=False)
reports2021.to_csv('../data/reports2021.csv', index=False)
```
